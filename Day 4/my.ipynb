```python
import math


def calculate_hypotenuse(base, height):

    if height > 0 and base > 0:
        hypotenuse = math.sqrt(base**2 + height**2)
        return hypotenuse
    else:
        return "Invalid input"


hypotenuse_value = calculate_hypotenuse(3, 4)
print(hypotenuse_value)
```

## Docstring & type hints


```python
def f1_score(precision: float, recall: float) -> float:
    """Return the harmonic mean of precision and recall."""

    if precision+recall == 0:
        return 0.0
    return 2*(precision*recall)/(precision+recall)


print(f1_score(0.8, 0.6))
```

```python
def average(*args):
    # print(args)
    # print(type(args))

    if not args:
        return 0.0

    return sum(args)/len(args)


print(average(10, 12, 50))
print(average(25, 80, 47, 30, 60))
```

```python
import math


class Circle:

    def __init__(self, radius):
        self.radius = radius    #  Instance Attribute

    def calculate_area(self):
        area = math.pi * self.radius**2
        return area


circle1 = Circle(7)
print(circle1.calculate_area())
```

```python
class RunningAverage:

    def __init__(self):
        self.total = 0
        self.count = 0

    def add(self, value):
        self.total += value
        self.count += 1

    def value(self):
        if self.count == 0:
            return 0.0
        return self.total / self.count


loss_tracker = RunningAverage()
for loss in [0.9, 0.7, 0.5]:
    loss_tracker.add(loss)
print(loss_tracker.value())
```
