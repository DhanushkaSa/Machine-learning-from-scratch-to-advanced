Question 04


```python
def get_priority(label: str, confidence: float) -> str:
    if confidence < 0.50 or label.strip().lower() == 'unknown':
        return "HIGH"
    elif 0.50 <= confidence < 0.80:
        return "MEDIUM"
    else:
        return "LOW"


print(get_priority('positive', 0.76))
```

```python
samples = [
    {'age': 22, 'label': 'ham', 'score': 0.91, 'country': 'LK',
        'is_premium': False, 'has_missing_values': False},
    {'age': 17, 'label': 'ham', 'score': 0.91, 'country': 'LK',
        'is_premium': False, 'has_missing_values': False},
    {'age': 30, 'label': 'spam', 'score': 0.88, 'country': 'US',
        'is_premium': True, 'has_missing_values': False},
    {'age': 40, 'label': 'ham', 'score': None, 'country': 'US',
        'is_premium': True, 'has_missing_values': False},
    {'age': 50, 'label': 'ham', 'score': 0.75, 'country': 'US',
        'is_premium': False, 'has_missing_values': False},
]


def is_eligible(samples_dict):

    eligible_count = 0

    for dictionary in samples_dict:
        if 18 < dictionary['age'] < 65 and dictionary['label'] != 'spam' and dictionary["score"] is not None and dictionary['score'] > 0.60 and (dictionary['country'] == 'LK' or dictionary['is_premium']) and not dictionary['has_missing_values']:
            eligible_count += 1
            print(True)
        else:
            print(False)

    return eligible_count


eligible_count = is_eligible(samples)
print(f'Eligible sample counts: {eligible_count}')
```

```python
scores = [0.91, 0.88, None, 0.95, 0.25, 0.72, 0.81]
valid_score = []

processed_count = 0

for score in scores:
    processed_count += 1
    if score is None:
        continue
    if score < 0.30:
        print("Critical anomaly detected")
        break

    valid_score.append(score)


print(f'Valid scores : {valid_score}')
print(f"Number of process items : {processed_count}")
```

Question 07


```python
attempt_results = [0.72, 0.81, 0.86, 0.93, 0.94]
threshold = 0.90
max_attempts = 10
attempt = 1
index = 0

while attempt <= max_attempts and index <= len(attempt_results):
    accuracy = attempt_results[index]
    print(f'Training attempt {attempt} accuracy {accuracy}')
    if accuracy >= threshold:
        print("training succeed")
        break
    attempt += 1
    index += 1
```

Question 08


```python
raw_values = ['0.91', 'bad', '  ', '0.72', '0.88x', '0.65', '']

clean_values = []

try:
    for value in raw_values:
        clean_value = value.strip()
        if clean_value in ('bad', ' ', '0.88x', ''):
            continue
       

        clean_values.append(float(clean_value))

    if not clean_values:
        print("No valid values")

    average = sum(clean_values)/len(clean_values)
    print(f'Average of clean values : {average:.3f}')
    print(clean_values)

except ValueError:
    print("No valid values")
```

Question 10

```python
predictions = ['cat','dog','cat','bird','dog']
true_labels = ['cat','cat','cat','bird','cat']
mistake = []

for index,label in enumerate(predictions):
    if label != true_labels[index]:
        mistake.append((index,label,true_labels[index]))

print(f'Mistake count : {len(mistake)}')
print(f'Mistake with the highest index : {max(mistake)}')

```
