## Python Function


```python
def calculate_accuracy(correct_labels, total_labels):
    accuracy = correct_labels/total_labels
    return accuracy


accuracy_value = calculate_accuracy(8, 10)

print(f'accuracy value {accuracy_value}')
```

```python
# Question
import math


def hypotenuse_triangle(base, height):

    if base <= 0 or height <= 0:
        return "Error base and height can't be negative"

    return math.sqrt(base**2 + height**2)


print(f'Hypotenuse value {hypotenuse_triangle(3, 4)}')
```

```python
def calculate_accuracy_2(correct_labels, total_labels=20):
    accuracy = correct_labels/total_labels
    return accuracy


accuracy_value = calculate_accuracy_2(8, 10)

print(f'accuracy value {accuracy_value}')
```

```python
def calculate_accuracy_3(correct_labels, total_labels):
    accuracy = correct_labels/total_labels
    return accuracy


accuracy_value = calculate_accuracy_3(total_labels=8, correct_labels=10)

print(f'accuracy value {accuracy_value}')
```

```python
def average(*args):

    print(args)
    print(type(args))

    if not args:
        return 0.0

    return sum(args)/len(args)


print(f'average value {average(1, 2, 3, 4, 5)}')
print(f'average value {average(1, 2, 3)}')
```

```python
def average_sample(**kwargs):
    print(kwargs)
    print(type(kwargs))

    if not kwargs:
        return 0.0

    return sum(kwargs.values())/len(kwargs.values())


print(f'average value {average_sample(num_1=8, num_2=3, num_3=2)}')
print(f'average value {average_sample(num_1=4, num_2=5)}')
```

## Docstrings


```python
import math


def calculate_area(radius):
    """
     Calculate the area of a circle given its radius.

     Args:
        radius (float): The radius of the circle.

     Returns:
        float: The calculated area
    """

    return math.pi * (radius**2)
```

```python
print(calculate_area.__doc__)
print(calculate_area(5))
```

## Lambda function


```python
def find_square(x): return x**2


print(find_square(5))
```

```python
my_list = [54, 12, 36, 47, 79, 20]

result = sorted(my_list, key=None, reverse=False)
print(result)
```

```python
students = [("John", 85), ("Alice", 92), ("Bob", 78)]

print(sorted(students))
```

```python
result = sorted(students, key=lambda x: x[1], reverse=False)
print(result)
```

```python
count = [80, 30, 10, 20, 78]
result_2 = sorted(count, key=lambda x: x, reverse=False)
print(result_2)
```

```python
users = [
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35}
]

result = sorted(users, key=lambda x: x["age"], reverse=False)

print(result)
```

```python
maximum = max(users, key=lambda x: x["age"])
print(f'Maximum age : {maximum}')
```

```python
minimum = min(users, key=lambda x: x["age"])
print(f'Minimum age : {minimum}')
```

## List & Dictionary Comprehensions


```python
my_list = [8, 5, 9, 4, 20]
squard_list = [x**2 for x in my_list]
print(squard_list)
```

```python
squard_list_greater = [x**2 for x in my_list if x > 5]
print(squard_list_greater)
```

```python
labels = ['cat','dog','fish','bear']

convert_dict = {label: index for index, label in enumerate(labels)}
print(convert_dict)
```
