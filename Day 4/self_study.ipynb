```python
raw_rows = [
    {'id': 'S001', 'label': 'cat', 'score': '0.95'},
    {'id': 'S002', 'label': 'dog', 'score': 'bad'},
    {'id': 'S003', 'label': 'cat', 'score': '0.82'},
    {'id': 'S004', 'label': 'bird', 'score': '0.61'},
]

valid_scores = []
label_counts = {}
unique_labels = set()

for row in raw_rows:
    label = row['label'].strip().lower()
    unique_labels.add(label)

    if label not in label_counts:
        label_counts[label] = 0
    label_counts[label] += 1

    try:
        score = float(row['score'])
        valid_scores.append(score)

    except ValueError:
        print(f'Skipping invalid score for sample {row['id']}')

average_score = sum(valid_scores)/len(valid_scores)

print(f'Unique labels : {unique_labels}')
print(f'Label counts : {label_counts}')
print(f'Average valid score : {average_score:.2f}')
```
