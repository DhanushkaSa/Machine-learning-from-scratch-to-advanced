```python
def predict(score, threshold):
    return 'positive' if score >= threshold else 'negative'


print(predict(threshold=0.8, score=0.9))
```

```python
def f1_score(precision: float, recall: float) -> float:
    """
    Return the harmonic mean of precision and recall.

    """

    if precision + recall == 0:
        return 0.0
    return 2*(precision*recall)/(precision+recall)


print(f1_score(0.8, 0.6))
```

```python
def average(*args):
    return sum(args)/len(args)


print(average(10, 20, 30))
print(average(25, 80))
print(average(10))
```

```python
samples = [
    {'id': 'a', 'score': 0.72},
    {'id': 'b', 'score': 0.95},
    {'id': 'c', 'score': 0.61},
]

ranked = sorted(samples, key=lambda row: row['score'], reverse=True)

for row in ranked:
    print(row['id'], row['score'])
```

```python
sample_max = max(samples, key=lambda row: row['score'])
print(sample_max)
```

```python
sample_min = min(samples, key=lambda row: row['score'])
print(sample_min)
```
