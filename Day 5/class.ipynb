Inheritance


```python
class Vehicle:
    has_wheels = True

    def display(self):
        print("This is vehicle")


class Car (Vehicle):

    def display(self):   # Method Overriding
        print("this is car")


car_1 = Car()
print(car_1.has_wheels)
car_1.display()
```

```python
class Vehicle:

    def __init__(self,manufacture_date,color):
        self.manufacture_date = manufacture_date
        self.color = color

class Car(Vehicle):
    
    def __init__(self,brand,manufacture_date,color):

        super().__init__(manufacture_date,color)
        self.brand = brand
        
        

    def display(self):
        print(f'Manufacuturing Date: {self.manufacture_date} \nColor: {self.color} \nBrand: {self.brand}')

car_1=Car("BMW",'2026-01-10','blue')
car_1.display()


```

Encapsulation

```python
class Car:
    def __init__(self,brand,color):
        self.brand=brand
        self.color = color

    def display(self):
        print(f'{self.brand}, {self.color} color car')

car_1=Car("BMW","Red")
print(car_1.brand)
print(car_1.color)                                                   # We can direct access attributes
```

```python
class Car:
    def __init__(self,brand,color):
        self._brand=brand
        self._color = color

    def _display(self):
        print(f'{self._brand}, {self._color} color car')

class CarSub(Car):
    
    def info(self):
        print(f'{self._brand}, {self._color} color car')

car_1=CarSub("BMW","Red")
car_1.info()                   # Don't recommend. this is a bad practice. Not give run time error 
```

protected attributes and methods

```python
class Car:
    def __init__(self,brand,color):
        self._brand=brand
        self._color = color

    def _display(self):
        print(f'{self._brand}, {self._color} color car')

class CarSub(Car):

    
    
    def info(self):
        super()._display()
        print(f'{self._brand}, {self._color} color car')

car_1=CarSub("BMW","Red")
car_1.info()                   # Don't recommend. this is a bad practice. Not give run time error 
```

private attributes and methods

```python
class Car:
    def __init__(self,brand,color):
        self.__brand=brand
        self.__color=color

    def __display(self):
        print(f'{self.__brand}, {self.__color} color car')

car_1=Car("Benz","Gray")
print(car_1.__brand)
print(car_1.__color)
car_1.__display()
```

```python
class Car:
    def __init__(self,brand,color):
        self.__brand=brand
        self.__color=color

    def __display(self):
        print(f'{self.__brand}, {self.__color} color car')

class ElectricCar(Car):

    def info_car(self):
        super().__display()
        print(f'{self.__brand}, {self.__color} color car')


car_1=ElectricCar("Benz","Gray")
car_1.info_car()
```

@property decorator

```python
class Car:
    def __init__(self,brand,color):
        self._brand=brand
        self._color=color


    
    def __display(self):
        print(f'{self._brand}, {self._color} color car')

    @property
    def get_brand(self):
        return self._brand

car_1=Car("Benz","Gray")
car_1.get_brand
```

Composition

```python
class Engine:
    def __init__(self,horsepower):
        self.horsepower=horsepower

    def start(self):
        return (f'Engine with {self.horsepower} HP is roaring to life!')
    
class Battery:
    def __init__(self,capacity):
        self.capacity=capacity

    def charge(self):
        return (f'Battery charging to {self.capacity} kwh...')
    

# The car class USES Engine and Battery (Composition)

class Car:
    def __init__(self,model,horse_power,charging_capacity):
        self.horse_power=Engine(horse_power)
        self.charging_capacity=Battery(charging_capacity)
        self.model=model


car_1=Car("Benz",5000,75)
print(car_1.horse_power.start())
print(car_1.charging_capacity.charge())
print(car_1.model)

```
