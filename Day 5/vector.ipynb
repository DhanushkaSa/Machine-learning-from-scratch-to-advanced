```python

new_vector = []


def vector_addition(vector_1, vector_2):

    if len(vector_1) != len(vector_2):
        return "Vector length must be equal"
    for value in range(len(vector_1)):
        new_vector.append(vector_1[value]+vector_2[value])
    return new_vector


print(vector_addition([20, 600, 70, 300, 200], [8, 69, 30, 75, 10]))
```

```python
def scalar_multiplication(scalar, vector):
    return [scalar*value for value in vector]


print(scalar_multiplication(2, [20, 50, 60, 70]))
```

```python
import math


def calculate_magnitude(vector):
    magnitude = 0
    for value in vector:
    
        magnitude = magnitude + (value**2)
    return math.sqrt(magnitude)


print(calculate_magnitude([1, -2, 2]))
```
