```python
# Question 2

raw_labels = [
    ' Positive ',
    'NEGATIVE',
    'neutral',
    'positive',
    ' Negative ',
    'pos',
    ' negative '
]

clean_labels = []
rejected_label_count = 0


for label in raw_labels:
    status = label.strip().lower()
    if status == 'positive':
        clean_labels.append(label)
    elif status == 'negative':
        clean_labels.append(label)
    else:
        rejected_label_count += 1


print(clean_labels)
print(f'Rejected label count : {rejected_label_count}')
```

```python
# Question 6

scores = [0.91, 0.88, None, 0.95, 0.25, 0.72, 0.81]
valid_score = []
number_of_process_items = 0

for score in scores:
    number_of_process_items += 1
    if score == None:
        continue
    if score < 0.30:
        print("Critical anomaly detected")
        break

    valid_score.append(score)

print(f"Number of process items : {number_of_process_items}")
print(f'Valid score : {valid_score}')
```

```python
# Question 8

raw_values = ['0.91', 'bad', ' ', '0.72', '0.88x', '0.65', '']

clean_values = []

for value in raw_values:
    if value in ('bad', ' ', '0.88x', ''):
        continue

    try:

        clean_values.append(float(value))
    except ValueError:
        print('No valid values')

print(clean_values)
print(f'Average of clean values : {sum(clean_values)/len(clean_values):.3f}')
```

```python
# Question 10

predictions = ['cat', 'dog', 'cat', 'bird', 'dog']
true_labels = ['cat', 'cat', 'cat', 'bird', 'cat']

mistakes = []

for index, predicted in enumerate(predictions):
    actual = true_labels[index]

    if predicted != actual:
        mistakes.append((index, predicted, actual))

print("Mistake count:", len(mistakes))
print("Mistake with the highest index:", mistakes[-1])
```

```python
# Question 11

results = [('cat', 0.97), ('dog', 0.62), ('bird', 0.88), ('cat', 0.75)]

maximum = results[0][1]
best_result = results[0]

for result in results:
    if result[1] >= 0.85:
        print(f'{result[0]} -> {result[1]}')

    if maximum <= result[1]:
        maximum = result[1]
        best_result = result

print(best_result)
```

```python
# Question 13

labels = ['cat', 'dog', 'cat', 'bird', 'dog', 'cat', 'bird', 'dog', 'dog']

label_counts = {}

for label in labels:
    if label in label_counts:
        label_counts[label] += 1
    else:
        label_counts[label] = 1

majority_label = max(label_counts, key=label_counts.get)

print(label_counts)
print("Majority label:", majority_label)
```
