```python
# Question 06

scores = [0.91,0.88,None,0.95,0.25,0.72,0.81]
valid_score=[]
count = 0

for score in scores:
    count+=1
    if score is None:
        continue
    if score<0.3:
        print("Critical anomaly detected")
        break
    

    
    valid_score.append(score)
print(valid_score)

print(f'Number of process items {count}')
    
```

```python
# Question 07

attempt_results = [0.72,0.81,0.86,0.93,0.94]
threshold = 0.90
max_attempts = 10
attempt = 1
index = 0


while attempt <= max_attempts and index<=len(attempt_results):
    accuracy = attempt_results[index]
    print(f'Training attempt {attempt} accuracy {accuracy} ')

    if accuracy >= threshold:
        print("training succeed")
        break
    attempt += 1
    index += 1




        


```

# Exception Handling

```python
x = 10
try:
    num=x/0
    print(num)
except ZeroDivisionError:
    print("Error : Division by zero")
```

```python
raw_values = ['0.83','0.49','bad','0.72']
clear_scores = []

for raw_value in raw_values:
    try:
        score = float(raw_value)
        clear_scores.append(score)
        
    except ValueError:
        print(f'Error : Invalid value {raw_value}')

print(clear_scores)

```
