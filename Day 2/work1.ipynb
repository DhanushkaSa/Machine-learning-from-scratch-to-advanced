```python
train_labels = ['cat','dog','bird','cat','dog']
test_labels = ['dog','fish','cat','fish','rabbit']

train_labels_set = set(train_labels)
test_labels_set = set(test_labels)

common_set = test_labels_set.difference(train_labels_set)
print(common_set)

duplicate_set=train_labels_set.intersection(test_labels_set)
print(duplicate_set)

is_duplicate_values = len(test_labels)!=len(test_labels_set)
print(is_duplicate_values)
```
